### Скрепинг таблиц

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_nuclear_weapons_tests').text
```

```python
# lxml - это парсер, словарь, с помощшью котрого BS понимает? что явялется тегами
soup = BeautifulSoup(url, 'lxml')
```

```python
# нашли 19 таблиц
len(soup.find_all('table'))
```

```python
tables = soup.find_all('table', {'class': "wikitable sortable"})
```

```python
# ищем ряды tr
rows = tables[1].find_all('tr')
```

```python
len(rows)
```

```python
rows[0]
# th - table header
```

```python
rows[0].find_all('th')[0].get_text().strip()
```

```python
# td - table data
# последний ряд тоже может быть table header
rows[-1]
```

```python
rows[1].find_all('td')[0].get_text().strip()
```

```python
dates = []

dates.append(rows[0].find_all('th')[0].get_text().strip())

for row in rows[1:]:
    dates.append(int(row.find_all('td')[0].get_text().strip().split()[-1]))
```

```python
print(dates)
```

```python
yields = []

for row in rows:
    try:
        yields.append(float(row.find_all('td')[1].get_text().strip()))
    except:
        yields.append(row.find_all('th')[1].get_text().strip())
print(yields)
```

```python
countries = []

for idx in range(len(rows)):
    if idx == 0:
        countries.append(rows[idx].find_all('th')[3].get_text().strip())
    else:
        countries.append(rows[idx].find_all('td')[3].get_text().strip())
        
print(countries)
```

```python
f = open('nuclear_tests.csv', 'w')
for idx in range(len(dates)):
    f.write(f'{dates[idx]},{yields[idx]},{countries[idx]}\n')
f.close()
```

```python
df = pd.read_csv('nuclear_tests.csv', index_col = None)
df.tail(20)
```

```python
unique_countries = set(countries)
unique_countries
```

```python
def average_yield(country):
    tests = []
    for idx in range(len(yields)):
        if countries[idx] == country:
            tests.append(yields[idx])
    return round(sum(tests) / len(tests), 2)
average_yield('Soviet Union')
```

```python
for country in set(countries[1:]):
    print(f'{country}: средняя можность взрыва {average_yield(country)} мегатон')
```

```python
tests = {}

for idx in range(1, len(countries)):
    tests[countries[idx]] = [countries.count(countries[idx])]
```

```python
def get_news(news0):
    soup0 = BeautifulSoup(requests.get(news0).text, 'lxml')
    author = soup0.find_all('meta', {'name': 'author'})[0].get('content')
    date = soup0.find_all('meta', {'itemprop': 'datePublished'})[0].get('content')
    text0 = [t.text for t in soup0.find_all('p', {'class': None})]
    final_text = ''.join(text0)
    final_text = final_text[:final_text.find(author)]
    final_text = final_text.replace('\xa0', '')
    return date, author, final_text
```

```python
from time import sleep

def get_uik_address(n_uik, reg):
    br.get('http://www.cikrf.ru/services/lk_address/?do=find_by_uik')
    uik_field = br.find_element_by_css_selector('#uik')
    uik_field.send_keys(n_uik)
    region_field = br.find_element_by_name('subject')
    region_field.send_keys(reg)
    button = br.find_element_by_link_text('Отправить запрос')
    button.click()
    try:
        
        first = br.page_source.find('Адрес помещения для голосования: ') + len('Адрес помещения для голосования: ')
        last = br.page_source.find('</p><p>Телефон помещения для голосования:')
        address  = br.page_source[first:last]
    except:
        address = 'Cannot find address'
    return address
```

```python
page = 1
link = f'http://www.biblio-globus.ru/search/catalog/products?query=Python&page={page}'
link
```

```python
all_titles = []
all_prices = []
for page in range(1,11):
    link = f'http://www.biblio-globus.ru/search/catalog/products?query=Python&page={page}'
    soup1 = BeautifulSoup(requests.get(link).text, 'lxml')
    titles = []
    for title in soup1.find_all('a', {'name'}):
        titles.append(title.get_text())
        prices = [price.get_text().split()[-2].replace(',', '.') for price in soup1.find_all('div',{'class': 'title_data price'})]
        all_titles.extend(titles)
        all_prices.extend(prices)
```
