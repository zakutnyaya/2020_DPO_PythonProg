```python
links = []
for link in soup.find_all('a', text = 'Summary'):
    links.append('https://www.hse.ru' + link.get('href'))
```

## xml

```python
# xml - это про струткуру хранения данных, а html - разметка
```

```python
group = """<group>
<number>134</number>
<student>
<firstname>Виталий</firstname>
<lastname>Иванов</lastname>
</student>
<student>
<firstname>Мария</firstname>
<lastname>Петрова</lastname>
</student>
</group>"""
```

```python
obj = BeautifulSoup(group, features = 'xml')
```

```python
# xml можно представить как дерево
print(obj.prettify())
```

```python
obj.group.student.lastname.text
```

```python
for student in obj.group.find_all('student'):
    print(student.lastname.text)
```

```python
import xml.etree.ElementTree as ET
```

```python
# парсим
tree = ET.fromstring(group)
```

```python
list(tree)
```

```python
for element in tree.findall('student'):
    print(element.find('lastname').text)
```

## API

```python
url = 'https://en.wikipedia.org/w/api.php'
params = {
    'action':'query',
    'list' : 'categorymembers',
    'cmtitle' : 'Category:Physics',
    'format' : 'xml'
}

g = requests.get(url, params=params)
```

```python
data = BeautifulSoup(g.text, features = 'xml')
```

```python
print(data.prettify())
```

```python
for cm in data.api.query.categorymembers('cm'):
    print(cm['title'])
```

```python
data.find('continue')['cmcontinue']
```

```python
params['cmcontinue'] = data.find('continue')[]
```

```python
from openpyxl import Workbook
from openpyxl.chart import BarChart, Reference

workbook = Workbook()
sheet = workbook.active

# Let's create some sample sales data
rows = [
    ["Product", "Online", "Store"],
    [1, 30, 45],
    [2, 40, 30],
    [3, 40, 25],
    [4, 50, 30],
    [5, 30, 25],
    [6, 25, 35],
    [7, 20, 40],
 ]

for row in rows:
    sheet.append(row)
```

```python
sheet.dimensions
```

```python
chart = BarChart()
```

```python
data = Reference(worksheet=sheet, min_row=1, max_row=8, min_col=2, max_col=3)
chart.add_data(data, titles_from_data=True)
sheet.add_chart(chart, 'E2')
workbook.save('chart.xlsx')
```
